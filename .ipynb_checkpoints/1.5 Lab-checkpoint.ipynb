```python
sales_channel = ('Web', 'Agent', 'Branch','Call Center')
responses = {'Yes': np.array([156, 666, 294,192]), 'No': np.array([1169, 2811, 2273,1573])}
width = 0.6  # the width of the bars: can also be len(x) sequence


fig, ax = plt.subplots()
bottom = np.zeros(4)

for rate, responses in responses.items():
    p = ax.bar(sales_channel, responses, width, label=rate, bottom=bottom)
    bottom += responses

    ax.bar_label(p, label_type='center')

ax.set_title('Response rate by the Sales Channel')
ax.legend()

plt.show()
```
